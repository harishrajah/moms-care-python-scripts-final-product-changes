```python
# columns  which are not null are used for predicting the value of 2013–14 column
year_df = df2[['2011–12','2012–13','2013–14']]

# dropping the 2013–14 column to get the training data
year_train_df = year_df.dropna(subset=['2013–14'],axis=0)
```

```python
# using linear regression
lm_fitting= LinearRegression()

# training the model with the training data for 2013–14
lm_fitting.fit(year_train_df[[x for x in year_train_df.columns if x !='2013–14']],year_train_df['2013–14'])

# value prediction
lm_fitting.predict(year_df.drop('2013–14',axis=1)) 

# assigning the predicted values to the null values of 2013–14 column
year_df['2013–14'][year_df['2013–14'].isnull()] = lm_fitting.predict(year_df.drop('2013–14',axis=1))

# rounding off values
year_df['2013–14'] = year_df['2013–14'].apply(lambda x : round(x*2)/2)
```

```python
year_df[year_df['2013–14'].isnull()] 
```

```python
# replacing missing values of 2013–14 column in year_df with our predicted data
df2= df2.combine_first(year_df) 
```

```python
# checking for null values
print(df2.isnull().sum())
```
